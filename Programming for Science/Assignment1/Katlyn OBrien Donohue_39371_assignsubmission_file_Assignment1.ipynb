```python
%matplotlib inline

# Will include the data points required to plot.
tvalues = []
Svalues = []
Ivalues = []
Rvalues = []

# initial population values
# s = Susceptible Population , i = Infected Population, r = Recovered Population
s = 45400
i = 2100
r = 2500

a = .00001
b = 1/14

# time in years
deltat     = .5
timePeriod = 0
endtime    = 10

# Adds initial values into respective lists.
tvalues.append(timePeriod)
Svalues.append(s)
Ivalues.append(i)
Rvalues.append(r)

while (timePeriod <=  endtime):
    # Sprime, Iprime, and Rprime are the rate of change of each population
    Sprime = (-a * s * i)
    Iprime = (a * s * i) - (b * i)
    Rprime = (b * i)
    s = s + (Sprime * deltat)
    i = i + (Iprime * deltat)
    r = r + (Rprime * deltat)
    timePeriod = timePeriod + deltat
    # appends the lists to include new values of the timePeriod, s, i, and r.
    tvalues.append(timePeriod)
    Svalues.append(s)
    Ivalues.append(i)
    Rvalues.append(r)

# Legend stuff from (http://matplotlib.org/1.3.0/examples/pylab_examples/legend_demo.html)
fig, ax = plt.subplots()
ax.plot (tvalues, Svalues, label='Susceptible')
ax.plot (tvalues, Ivalues, label='Infected')
ax.plot (tvalues, Rvalues, label='Recovered')
legend = ax.legend(loc='upper right', shadow=True)
xlabel('Time')
ylabel('Population')
```
